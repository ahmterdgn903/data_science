```python
def increase_pop(start_num,percent,num_of_day):
    pop = start_num
    pop_list = []
    for i in range(num_of_day):
        if i == 0:
            pop = start_num
            pop_list.append(pop)
        else:
            pop += pop*percent/100
            pop_list.append(pop)
    for index,i in enumerate(pop_list):
        print(index,i)
start_num = int(input('pls input a start number :'))
percent = int(input('pls input a average :'))
num_of_day = int(input('pls input a total days :'))
increase_pop(start_num,percent,num_of_day)
```

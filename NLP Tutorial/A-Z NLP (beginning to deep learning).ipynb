### Tokens

#### Sent Tokenize / Cumle Tokenlestirme

```python
# cumle tokenestirme/sent tokenize
# gerekli olan paketi ekleyelim
from nltk.tokenize import sent_tokenize
```

```python
# yapay zekada dogal dil isleme hakkinda bir metin wikipediadan
text = "Gelecekte, konuşma sentezleyiciler ve konuşma anlama alanındaki gelişmeler ve makine-insan iletişiminin gelişmesi, insanın makineden beklentilerini yükseltecektir. İnsanlar makinelerin kendisini anlamalarını isteyecek, karmaşık kullanımı olan makineler pazar bulamayacaktır. Giderek gelişen ve insanı anlayan makinelerin daha zeki olması insanın yaşam kalitesini yükselteceğinden, vazgeçilmez olması kaçınılmazdır. Zeki makine kavramı, yapay zekâ çalışmalarının hızlanmasına yol açmıştır. Geleceğin en önemli sektörlerinden biri olan yapay zekâ ile insanın iletişim kuracağı tek araç dildir."

sentences = sent_tokenize(text)
sentences
```

### Stopwords

Ingilizce dili icin orenek verirsek, to be yani 'am', 'is', 'are','was', 'were' vb. Bu ek fiiller hemen hemen her cumlede bulunur ama bir duygu ifade etmez. Peki buna neden ihtiyacimiz var? Su asamada olmasada bir sentiment analysis/duygu analizi calismasi yaparken bu tarz kelimelerin model icinde yer almasi modelin yanlis seyler ogrenmesine sebep olacakir.

```python
# oncelikle paketi dahil edelim
from nltk.corpus import stopwords
```

```python
# nltk corpus icin bircok dilde stopwords bulunmakta
stopwords.words('turkish')
```

```python
stopwords.words('english')
```

```python
# veri bilimi ustalarindan Andrew NG hakkinda wikipediadan bir metin
text = "Ng is an adjunct professor at Stanford University (formerly associate professor and Director of its AI Lab). Also a pioneer in online education, Ng co-founded Coursera and deeplearning.ai.[4] With his online courses, he has successfully spearheaded many efforts to 'democratize deep learning' teaching over 2.5 million students through his online courses.[5][2] He is one of the world's most famous and influential computer scientists being named one of Time magazine's 100 Most Influential People in 2012, and Fast Company's Most Creative People in 2014. Since 2018 he launched and currently heads AI Fund, initially a $175-million investment fund for backing artificial intelligence startups. He has founded Landing AI, which provides AI-powered SaaS products and Transformation Program to empower enterprises into cutting-edge AI companies.[6]"

#stopwords kelimeler icerdigi icin metun uzerinde islem yapmak istiyorsak once metnimizi kelimelere bolelim.
words = word_tokenize(text)
print('stopwords islemi oncesi :', words)
len(words)
```

```python
cleaned_words = [word for word in words if word not in StopWords]
print('stopwords kelimeleri silindikten sonra :', cleaned_words)
len(cleaned_words)
```

```python
ps = PorterStemmer()
words = ["python","pythoner","pythoning","pythoned","pythonly"]
[print(ps.stem(word)) for word in words]
```

```python
# birkac ornek verelim
# asagidaki orneklerde de gorecginiz gibi kelime turu verek de yapabiliriz.

print(lemmatizer.lemmatize('increases'))
print(lemmatizer.lemmatize('playing', pos="v"))
print(lemmatizer.lemmatize('playing', pos="v")) 
print(lemmatizer.lemmatize('playing', pos="n")) 
print(lemmatizer.lemmatize('playing', pos="a")) 
print(lemmatizer.lemmatize('playing', pos="r"))
print(lemmatizer.lemmatize("cats"))
print(lemmatizer.lemmatize("cacti"))
print(lemmatizer.lemmatize("geese"))
print(lemmatizer.lemmatize("rocks"))
print(lemmatizer.lemmatize("python"))
print(lemmatizer.lemmatize("better", pos="a"))
print(lemmatizer.lemmatize("best", pos="a"))
print(lemmatizer.lemmatize("run"))
print(lemmatizer.lemmatize("run",'v'))
```

### Named Entity Recognation (NER)

Named Entity Recognation (NER) turce anlami ile isimlendirilmis varlik cikarma anlamina geliyor. Metin icerisinde isim, oranizasyon, lokaysyon, yer ismi vb gibi varlik isimlerini bulunmasini sagliyor.

```python
# gerekli paketi dahil edelim.
text = "Space Exploration Technologies Corp., trading as SpaceX, is a private American aerospace manufacturer and space transportation services company headquartered in Hawthorne, California. It was founded in 2002 by Elon Musk with the goal of reducing space transportation costs to enable the colonization of Mars.[7][8][9] SpaceX has developed several launch vehicles and the Dragon spacecraft."
# ilk once tokenlestirme 
words = word_tokenize(text)
# POS taglama 
tags = nltk.pos_tag(words)
#sonra NER
ned = nltk.ne_chunk(tags)
```

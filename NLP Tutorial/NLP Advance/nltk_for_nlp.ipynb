### stopwords 

All Natural Languages have some words for making sentence good. Example a, an, the, end.... But in data analyasis for machine those are words are not any meanning. 

```python
from nltk.corpus import stopwords
```

```python
# check stopwords
stop_words = stopwords.words('English')
```

```python
words = word_tokenize(text)
```

```python
# extract stopwords from words
filtred_words = [word for word in words if word not in stop_words]
```

```python
filtred_words
# no stopwords
```

### Stemming

Finding stem of words

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
words = ['drive','driving','driver','drives','drove','cats','children']
```

```python
[ps.stem(word_stem) for word_stem in words]
```

### Part of Spech

The elements of the sentence

```python
tokens = word_tokenize(text)
```

```python
import nltk
```

```python
nltk.pos_tag(tokens)
```

### Named Entity Recognation

Person, company,...

```python
# after part of spech
named_ent_rec = nltk.ne_chunk(nltk.pos_tag(tokens))
```

```python
named_ent_rec
```

### Lemmtizing

```python
from nltk.stem import WordNetLemmatizer
```

```python
lem = WordNetLemmatizer()
```

```python
words = ['drive','driving','driver','drives','drove','cats','children']
```

```python
[lem.lemmatize(lem_word) for lem_word in words]
```
